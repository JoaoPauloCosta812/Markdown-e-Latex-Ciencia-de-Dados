### Fórmula de Bhaskara
A solução para uma função de segundo grau depende das suas raízes(valores de x). Como já vimos, coeficientes precisam ser números reais e angular diferente de zero. Sendo assim, temos a seguinte fórmula de Bhaskara:
$$ x = \frac{-b \pm \sqrt{\Delta}}{2a} $$

</div>
<br>                                                                    
Para melhor compreensão o seu cálculo é dividido em duas partes: discriminante da equação e operações para determinar as raízes.

**Discriminante**

A figura dentro da raiz na formula de Bhaskara é nomeada de discriminante. Seu símbolo é a letra grega delta e apresenta a determinada formula:
$$ \Delta = b^2 - 4 . a . c$$
<div align="center">
Fórmula da discriminante
</div>
<br>

- Se o delta for maior que zero, a equação terá dois valores reais e distintos.
- Se o delta for igual a zero, a equação terá somente um valor real ou dois resultados iguais.
- Se o delta for menor que zero, a equação não possuirá valores reais.
